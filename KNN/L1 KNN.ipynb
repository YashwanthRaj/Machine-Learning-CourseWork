```python
# Importing pands library 

import pandas as pd
```

```python
# Reading the data and storing the data in variable called "data", then changing the date from normal string to date format

data = pd.read_csv('weather.csv', parse_dates=True, index_col=0)
```

```python
# Visualizing column names and dataset

data.head()
```

```python
# Visualizing column indexes

data.index
```

```python
# Taking only required columns from uploaded dataset and saving in new variabe name

dataset = data[['Humidity3pm', 'Pressure3pm', 'RainTomorrow']]
```

```python
dataset.head()
```

```python
# Importing math plot library for graphs and stuff

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Plotting the graph, below "=="No"" will show all points where rain has not occured
fig, ax = plt.subplots()

dataset[dataset['RainTomorrow']=='No'].plot.scatter(x="Humidity3pm", y="Pressure3pm", c='b', alpha =0.25, ax=ax)
dataset[dataset['RainTomorrow']=='Yes'].plot.scatter(x="Humidity3pm", y="Pressure3pm", c='r', alpha =0.25, ax=ax)
```

```python
# Importing libraries required for KNN

import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
# Cleaning the dataset

dataset_clean = dataset.dropna()
len(dataset), len(dataset_clean)
```

```python
# Splitting the determining values from label
X = dataset_clean[['Humidity3pm', 'Pressure3pm']]
y = dataset_clean['RainTomorrow']

# Changing the values of dataset lable into 0 and 1
y = np.array([0 if value == 'No' else 1 for value in y])
```

```python
y
```

```python
# Splitting the dataset into training and testing 
# random state for randomly selecting the values 

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
# Model Building
neigh = KNeighborsClassifier()
neigh.fit(X_train, y_train)

# Testing the accuracy
y_pred = neigh.predict(X_test)
accuracy_score(y_test, y_pred)
```

```python
# Times when it is not raining 
sum(y == 0)/len(y)
```

```python
# Times when it is raining 
sum(y == 1)/len(y)
```

```python
X_map = np.random.rand(10000, 2)
X_map = X_map*(100, 50) + (0, 990)
```

```python
fig, ax = plt.subplots()

y_map = neigh.predict(X_map)

ax.scatter(x=X_map[:,0], y=X_map[:,1], c=y_map, alpha=.25)
```

```python
# Finding the values of K

scores = []
for k in range(1, 100):
    neigh = KNeighborsClassifier(n_neighbors=k)
    neigh.fit(X_train, y_train)
    y_pred = neigh.predict(X_test)
    score = accuracy_score(y_test, y_pred)
    scores.append(score)
```

```python
# Plot the scores, Now from here we can find K value

fig, ax = plt.subplots()

ax.plot(scores)
```

```python
# From the above graph we can Infer that K value can be 4 or 6 to achieve the maximum accuracy
```
